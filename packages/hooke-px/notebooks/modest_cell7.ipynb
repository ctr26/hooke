```python
from pathlib import Path
import polars as pl
import numpy as np
from dataclasses import dataclass
import json
from typing import Dict, Any


COMPOUND_CONTROLS: set[str] = {
    "REC-0000058",
    "REC-0000069",
    "REC-0000098",
    "REC-0000219",
    "REC-0000370",
    "REC-0000375",
    "REC-0000391",
    "REC-0000493",
    "REC-0000584",
    "REC-0000590",
    "REC-0000671",
    "REC-0000706",
    "REC-0000737",
    "REC-0000741",
    "REC-0000773",
    "REC-0000842",
    "REC-0000854",
    "REC-0000867",
    "REC-0000889",
    "REC-0000959",
    "REC-0000970",
    "REC-0001029",
    "REC-0001106",
    "REC-0001148",
    "REC-0001183",
    "REC-0001214",
    "REC-0001246",
    "REC-0001249",
    "REC-0001250",
    "REC-0001258",
    "REC-0001289",
    "REC-0001290",
    "REC-0001610",
    "REC-0001619",
    "REC-0001655",
    "REC-0001721",
    "REC-0001730",
    "REC-0001744",
    "REC-0001907",
    "REC-0002007",
    "REC-0002082",
    "REC-0002101",
    "REC-0002178",
    "REC-0003420",
    "REC-0003935",
    "REC-0003980",
    "REC-0003982",
    "REC-0004011",
    "REC-0004273",
    "REC-0004419",
}

NEGATIVE_CONTROLS = [
    "REC-GRNA-0009335",
    "REC-GRNA-0010053",
    "REC-GRNA-0010043",
    "REC-GRNA-0010071",
    "REC-GRNA-0009367",
    "REC-GRNA-0014046",
    "REC-GRNA-0010050",
    "REC-GRNA-0009413",
    "REC-GRNA-0009427",
    "REC-GRNA-0010039",
    "REC-GRNA-0010038",
    "REC-GRNA-0014049",
    "REC-GRNA-0010063",
    "REC-GRNA-0009337",
    "REC-GRNA-0010048",
    "REC-GRNA-0014047",
    "REC-GRNA-0010069",
    "REC-GRNA-0014051",
    "REC-GRNA-0014042",
    "REC-GRNA-0010036",
    "REC-GRNA-0014045",
    "REC-GRNA-0010044",
    "REC-GRNA-0010049",
    "REC-GRNA-0010056",
    "REC-GRNA-0010051",
    "REC-GRNA-0014052",
    "REC-GRNA-0009319",
    "REC-GRNA-0014050",
    "REC-GRNA-0009481",
    "REC-GRNA-0010054",
    "REC-GRNA-0014044",
    "REC-GRNA-0014048",
    "REC-GRNA-0014043",
    "REC-GRNA-0009383",
    "REC-GRNA-0009435",
    "REC-GRNA-0010066",
    "REC-GRNA-0009345",
    "REC-GRNA-0010041",
    "REC-GRNA-0010065",
    "REC-GRNA-0010052",
    "REC-GRNA-0009329",
    "REC-GRNA-0010068",
    "REC-GRNA-0010040",
    "REC-GRNA-0009503",
    "REC-GRNA-0010046",
    ### SIRNA
    "REC-SIRNA-047139",
    "REC-SIRNA-046347",
    "REC-SIRNA-023618",
    "REC-SIRNA-011891",
    "REC-SIRNA-070391",
    "REC-SIRNA-033322",
    "REC-SIRNA-059857",
    "REC-SIRNA-055380",
    "REC-SIRNA-013118",
    "REC-SIRNA-069541",
    "REC-SIRNA-041233",
    "REC-SIRNA-010840",
    "REC-SIRNA-040984",
    "REC-SIRNA-052968",
    "REC-SIRNA-067904",
    "REC-SIRNA-055333",
    "REC-SIRNA-048898",
    "REC-SIRNA-013779",
    "REC-SIRNA-056649",
    "REC-SIRNA-000616",
    "REC-SIRNA-021198",
    "REC-SIRNA-043501",
    "REC-SIRNA-058443",
    "REC-SIRNA-036731",
    "REC-SIRNA-062542",
    "REC-SIRNA-032472",
    "REC-SIRNA-059495",
    "REC-SIRNA-059760",
    "REC-SIRNA-016146",
    "REC-SIRNA-038275",
    "REC-SIRNA-069619",
    "REC-SIRNA-013244",
    "REC-SIRNA-042015",
    "REC-SIRNA-012441",
    "REC-SIRNA-006388",
    "REC-SIRNA-055947",
    "REC-SIRNA-035097",
    "REC-SIRNA-061483",
    "REC-SIRNA-038396",
    "REC-SIRNA-035423",
    "REC-SIRNA-028005",
    "REC-SIRNA-008605",
    "REC-SIRNA-038221",
    "REC-SIRNA-016879",
    "REC-SIRNA-039836",
    "REC-SIRNA-004205",
    "REC-SIRNA-055086",
    "REC-SIRNA-061270",
    "REC-SIRNA-012679",
    "REC-SIRNA-042894",
    "REC-SIRNA-042662",
    "REC-SIRNA-027850",
    "REC-SIRNA-063815",
    "REC-SIRNA-025061",
    "REC-SIRNA-067581",
    "REC-SIRNA-041265",
    "REC-SIRNA-000791",
    "REC-SIRNA-028754",
    "REC-SIRNA-047032",
    "REC-SIRNA-021964",
    "REC-SIRNA-037754",
    "REC-SIRNA-039652",
    "REC-SIRNA-037864",
]


class Tokenizer:
    def __init__(self):
        self.token_to_id = {}
        self.id_to_token = []

    def fit(self, df: pl.Series):
        unique_tokens: pl.Series = df.sort()
        for i, token in enumerate(unique_tokens):
            self.token_to_id[token] = i
            self.id_to_token.append(token)
        return self

    def transform(self, x):
        if isinstance(x, str):
            x = [x]
        return [self.token_to_id[token] for token in np.array(x).flatten()]
    
    def __len__(self):
        return len(self.id_to_token)

    def __call__(self, x):
        return self.transform(x)


CROSS_CELL_LINE_BASE_PATH = Path(
    "/rxrx/data/valence/phenomics/cross_cell_line__brightfield__pretrain__v1_1" # v1_0 or v1_1
)

RESERVATION_PATH = Path(
    "/rxrx/data/valence/internal_benchmarking/vcds1/v1_3_reservation.json"
)

CASE_STUDY_MOLECULES = Path("/rxrx/data/user/ali.denton/outgoing/candidates_for_case_studies.csv")

CROSS_CELL_LINE_OBS_PATH: Path = (
    CROSS_CELL_LINE_BASE_PATH
    / "cross_cell_line__brightfield__pretrain__v1_1_obs.parquet"
)
INTERNAL_BENCHMARKING_BASE_PATH = Path("/rxrx/data/valence/internal_benchmarking/")


DRUGSCREEN_OBS_PATH: Path = (
    INTERNAL_BENCHMARKING_BASE_PATH
    / "vcds1"
    / "drugscreen__cell_paint__v1_2"
    / "drugscreen__cell_paint__v1_2_obs.parquet"
)
CROSS_CELL_LINE_TEST_OBS_PATH: Path = (
    INTERNAL_BENCHMARKING_BASE_PATH
    / "vcds1"
    / "cross_cell_line__brightfield__v1_1"
    / "cross_cell_line__brightfield__v1_1_obs.parquet"
)

# COMMON_COLUMNS: list[str] = [
#     "path",
#     "assay_type",
#     "shape",
#     "source",
#     "assay_type_id",
#     "experiment",
#     "experiment_id",
#     "cell_condition",
#     "cell_condition_id",
#     "comp_id",
#     "c_concs",
# ]

COMPOUND_PATTERN = r"^REC-\d+$"
TARGET_WELLS = 1000
TRAIN_FRACTION = 0.9
METADATA_SPLIT_SEED = 123456
OBS_SPLIT_SEED = 123456789


# def assign_random_split(
#     df: pl.DataFrame, train_fraction: float, seed: int
# ) -> pl.DataFrame:
#     """Assign a reproducible train/val split."""
#     n_train = int(df.height * train_fraction)
#     return df.with_columns(
#         pl.when(pl.int_range(0, pl.len()).shuffle(seed=seed) < n_train)
#         .then(pl.lit("train"))
#         .otherwise(pl.lit("val"))
#         .alias("split")
#     )

# Images with zarrs that are known to be corrupted
SKIP_IMAGES: list[str] = pl.read_csv(
    "/mnt/ps/home/CORP/jason.hartford/project/big-x/joint-model/metadata/zarr_scan_shards/skip_paths_all.csv"
)["path"].to_list()

@dataclass(frozen=True)
class SplitIndices:
    finetune: np.ndarray
    validation: np.ndarray
    test: np.ndarray

def load_vcb_split_indices(splits_path: Path) -> tuple[Dict[str, Any], SplitIndices]:
    """Load finetune/validation/test split indices."""
    with splits_path.open() as handle:
        splits_json = json.load(handle)
    first_fold = splits_json["folds"][0]
    split_indices = SplitIndices(
        finetune=np.array(first_fold["finetune"]),
        validation=np.array(first_fold["validation"]),
        test=np.array(first_fold["test"]),
    )
    return splits_json, split_indices

# Load the DRUGSCREEN splits
SPLITS_PATH: Path = (
    INTERNAL_BENCHMARKING_BASE_PATH
    / "vcb"
    / "splits"
    / "drugscreen__cell_paint__v1_2"
    / "split_compound_random__v1.json"
)
_, vcb_splits = load_vcb_split_indices(SPLITS_PATH)

CROSS_CELL_LINE_SPLITS_PATH: Path = (
    INTERNAL_BENCHMARKING_BASE_PATH
    / "vcb"
    / "splits"
    / "cross_cell_line__brightfield__v1_1"
    / "split_cell_type_manual__v1.json"
)

_, cross_cell_line_splits = load_vcb_split_indices(CROSS_CELL_LINE_SPLITS_PATH)
```

```python
import polars as pl

def get_held_out_rec_ids(df, cell_type, sample_fraction=0.1):
    """
    Samples unique individual IDs (atoms) from the rec_id lists 
    for a specific cell type.
    """
    return (
        df.filter(pl.col('cell_type') == cell_type)
          .select(pl.col('rec_id'))
          .explode(pl.col('rec_id'))  # Flatten ['A', 'B'] into 'A', 'B'
          .unique(maintain_order=True)                   # Get unique pool of rec_ids
          .sample(fraction=sample_fraction, shuffle=True, seed=12345)
          .to_series()
          .to_list()                  # Return as Python list for use in filters
    )

# 1. Get the list of individual IDs (genes/compounds) to strictly hold out
arpe_held_out = get_held_out_rec_ids(df, 'ARPE19')
pl.DataFrame(arpe_held_out).write_parquet('/mnt/ps/home/CORP/jason.hartford/project/big-x/big-img/metadata/arpe_held_out.parquet')
rptec_held_out = get_held_out_rec_ids(df, 'RPTEC')
pl.DataFrame(rptec_held_out).write_parquet('/mnt/ps/home/CORP/jason.hartford/project/big-x/big-img/metadata/rptec_held_out.parquet')
huvec_held_out = get_held_out_rec_ids(df, 'HUVEC_TNFalpha')
pl.DataFrame(huvec_held_out).write_parquet('/mnt/ps/home/CORP/jason.hartford/project/big-x/big-img/metadata/huvec_held_out.parquet')

# 2. Assign split based on overlap (intersection)
# If a row contains ANY held-out ID, it goes to validation.
df = df.with_columns(
    pl.when(
        (pl.col("cell_type") == 'ARPE19') & 
        (pl.col("rec_id").list.set_intersection(arpe_held_out).list.len() > 0)
    ).then(pl.lit("valid_arpe19"))
    
    .when(
        (pl.col("cell_type") == 'RPTEC') & 
        (pl.col("rec_id").list.set_intersection(rptec_held_out).list.len() > 0)
    ).then(pl.lit("valid_reptec"))
    
    .when(
        (pl.col("cell_type") == 'HUVEC_TNFalpha') & 
        (pl.col("rec_id").list.set_intersection(huvec_held_out).list.len() > 0)
    ).then(pl.lit("valid_huvec_tnfalpha"))
    
    .otherwise(pl.col("split")) # Preserves existing 'split' value (e.g., 'train')
    .alias("split")
)
```

# Active Learning Data

```python
import zarr
import pickle
import polars as pl

def get_shape(path) -> tuple[int, int, int]:
    if len(path) > 0:
        return tuple(zarr.open(str(path), mode='r').shape)
    else:
        return None

al = pl.read_parquet("/mnt/ps/home/CORP/jason.hartford/pythondev/photosynthetic/notebooks/AL_metadata.parquet")
al = al.with_columns(
    pl.lit("train").alias("split"),
    n_perts = pl.col("rec_id").list.len(),
    rec_id_string = pl.col('rec_id').list.sort().list.join('-'),
    pert_type_string = pl.col("perturbation_type").list.sort().list.join('-')
).with_row_index("row_id")
al = al.with_columns(
        pl.when(pl.col("n_perts") > 1)
        .then(pl.lit("valid_al"))
        .otherwise(pl.lit("train"))
        .alias("split")
    )
pert_types = ["COMPOUND-GUIDE", "COMPOUND-COMPOUND", "GUIDE-GUIDE"]
for pert_type in pert_types:
    train_ids = al.filter((pl.col("n_perts") > 1) & (pl.col("pert_type_string") == pert_type))['rec_id_string'].unique(maintain_order=True).sample(n=250, seed=420, shuffle=True)
    al = al.with_columns(
        pl.when(pl.col("rec_id_string").is_in(train_ids))
        .then(pl.lit("train"))
        .otherwise(pl.col("split"))
        .alias("split")
    )
# see code below for how this was generated
shapes = pickle.load(open("shapes.pkl", "rb"))
al = al.with_columns(
    shape = pl.Series(shapes)
)

al = al.filter(pl.col("shape") == [2048, 2048, 3]).with_columns(
    pl.lit("brightfield_3channel").alias("assay_type"),
    pl.lit("active_learning").alias("source"),
    concentration = (pl.col("concentration") * 1000).list.eval(
        pl.element().cast(pl.String),
    )
).rename({
    "path": "image_path",
    "experiment": "experiment_label",
    "address": "well_address"
}).with_columns(
    pl.concat_str(['plate', 'order_id' ,'read_id'], separator='_').alias('plate_order_read_id'),
)

stats = (
    al
    .group_by(["n_perts", "pert_type_string", "split"])
    .agg(
        pl.col("rec_id_string").n_unique().alias("unique_count"),
        pl.len().alias("total_rows")
    )
    .sort(["n_perts", "pert_type_string", "split"])
)
with pl.Config(tbl_rows=100):
    print(stats)
```
